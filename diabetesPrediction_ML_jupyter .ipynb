```python
X=df[['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age']].values
Y=df['Outcome'].values
X,Y
```

# Training& Testing the model

# Training:


# Splitting the data into Testing and Training sets:

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,random_state = 99)
x_train,x_test,y_train,y_test
#To get the same split every time, we can use the random_state attribute.
#We choose an arbitrary number to give it, and then every time we run the code, we will get the same split.
```

```python
#choosing the model
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(solver='liblinear')
model.fit(x_train,y_train)
```

```python
print("Testing Set :",x_test.shape)
print("Training Set : ",x_train.shape)
print("Whole data set : ",X.shape)
```

```python
y_pred = model.predict(x_test)
y_pred,y_test
```

# Testing

```python
print("Count of correct predictions:")
(y_pred==y_test).sum()
```

```python
print("Total :")
len(y_test)
```

```python
# Therefore, accuracy is:
149/192 
```

```python
model.score(x_test,y_test)
```

# Evaluation Metrics

# Accuracy , recall , precision & f1score:

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
print("Accuracy : " ,accuracy_score(y_test,y_pred))
print("Precision Score : ",precision_score(y_test,y_pred))
print("Recall Score : ",recall_score(y_test,y_pred))
print("f1Score : " , f1_score(y_test,y_pred))
```

```python
pd.DataFrame({'Actual':y_test[0:51] , 'Prediction':y_pred[0:51] })
#Printing and comparing the Actual and predicted values of the first 50 
```

```python
#Till now our model has just predicted 0 or 1 but we can also predict the actual probability
y_pred_proba  = model.predict_proba(x_test)
y_pred_proba
```

```python
y_pp=[]
for i in  range(len(y_pred_proba)):
    y_pp.append(y_pred_proba[i][0])
pd.DataFrame({'Actual':y_test[0:51] , 'Prediction':y_pp[0:51] })
```

# Getting the Confusion Matrix


```python
from sklearn.metrics import confusion_matrix
cmat = confusion_matrix(y_test,y_pred)
#confusionMatrix povides us with 4 values
#1-TRUE POSITIVE: predicted positive and actually positive.
#2-TRUE NEGATIVE: preidcted negative and actually negative.
#3-FALSE POSITIVE: preidcted positive but actually negative.
#4-FALSE NEGATIVE: predicted negative but actually positive.
sns.heatmap(cmat,annot = True,fmt="d",cmap='Blues' )
#plt.show()
cmat
#(0,0) - True Negative
#(0,1) - False positive
#(1,0) - False Negative
#(1,1) - True positive
```
